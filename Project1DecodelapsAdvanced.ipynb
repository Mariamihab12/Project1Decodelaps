```python
import random
import datetime
import nltk
from nltk.tokenize import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer

# Download required resources
nltk.download('punkt')
nltk.download('vader_lexicon')

class SmartChatBot:

    def __init__(self, name="AI_Bot"):
        self.name = name
        self.memory = []
        self.user_name = ""
        self.sia = SentimentIntensityAnalyzer()

        self.intents = {
            "greeting": ["hi", "hello", "hey"],
            "bye": ["bye", "exit", "quit"],
            "thanks": ["thanks", "thank you"],
            "help": ["help"],
        }

    def analyze_sentiment(self, text):
        score = self.sia.polarity_scores(text)

        if score['compound'] >= 0.5:
            return "positive"

        elif score['compound'] <= -0.5:
            return "negative"

        else:
            return "neutral"

    def get_response(self, user_input):

        tokens = word_tokenize(user_input.lower())

        # Store conversation memory
        self.memory.append(user_input)

        # Greeting
        if any(word in tokens for word in self.intents["greeting"]):
            return f"Hello {self.user_name if self.user_name else ''}! How can I help you today?"

        # Goodbye
        elif any(word in tokens for word in self.intents["bye"]):
            return "Goodbye! Have a great day!"

        # Thanks
        elif any(word in tokens for word in self.intents["thanks"]):
            return "You're welcome!"

        # Help
        elif "help" in tokens:
            return """
I can:
- Chat with you
- Detect sentiment
- Remember your name
- Tell date/time
- Do calculations
- Store conversation history
            """

        # Save user name
        elif "my" in tokens and "name" in tokens:
            try:
                index = tokens.index("is")
                self.user_name = tokens[index + 1]
                return f"Nice to meet you, {self.user_name}!"
            except:
                return "I couldn't understand your name."

        # Time
        elif "time" in tokens:
            current_time = datetime.datetime.now().strftime("%H:%M:%S")
            return f"Current time is {current_time}"

        # Date
        elif "date" in tokens:
            current_date = datetime.datetime.now().strftime("%Y-%m-%d")
            return f"Today's date is {current_date}"

        # Calculator
        elif "calculate" in tokens:
            try:
                expression = user_input.replace("calculate", "")
                result = eval(expression)
                return f"The result is {result}"
            except:
                return "Invalid mathematical expression."

        # Sentiment Analysis
        sentiment = self.analyze_sentiment(user_input)

        if sentiment == "positive":
            return "You seem happy today!"

        elif sentiment == "negative":
            return "I'm sorry you're having a bad day."

        else:
            neutral_responses = [
                "Interesting!",
                "Tell me more.",
                "I understand.",
                "Can you explain further?"
            ]
            return random.choice(neutral_responses)

    def show_memory(self):
        print("\nConversation History:")
        for i, msg in enumerate(self.memory, 1):
            print(f"{i}. {msg}")

# Run chatbot
bot = SmartChatBot()

print(f"{bot.name}: Hello! I am your smart AI chatbot.")
print("Type 'bye' to exit.\n")

while True:

    user_input = input("You: ")

    response = bot.get_response(user_input)

    print(f"{bot.name}: {response}")

    if user_input.lower() in ["bye", "exit", "quit"]:
        break

# Show conversation history at end
bot.show_memory()
```

This project is an advanced rule-based AI chatbot developed using Python and NLP techniques.

The chatbot can:
- Understand user input using Natural Language Processing (NLP)
- Detect user sentiment (positive, negative, neutral)
- Respond intelligently based on predefined intents
- Remember user information during conversation
- Perform mathematical calculations
- Provide date and time information
- Store conversation history

Technologies Used:
- Python
- NLTK (Natural Language Toolkit)
- Sentiment Analysis
- Object-Oriented Programming (OOP)

The main objective of this project is to simulate intelligent human-computer interaction using NLP concepts and conversational AI techniques.
